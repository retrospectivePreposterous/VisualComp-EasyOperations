```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = [10, 10]
```

```python
img = cv2.imread('pato.png') #Por defecto opencv lee en BGR
wb = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(wb,cmap="gray") 
```

```python
sns.heatmap(wb, annot=True, fmt="d", cmap="gray", cbar=False,
           xticklabels=False, yticklabels=False)
```

```python
inv = np.copy(wb)

inv = 255-inv
sns.heatmap(inv, annot=True, fmt="d", cmap="gray", cbar=False,
           xticklabels=False, yticklabels=False)
```

```python
thres = np.copy(wb)
thres[thres < 92] += 50


sns.heatmap(thres, annot=True, fmt="d", cmap="gray", cbar=False,
           xticklabels=False, yticklabels=False)
```

```python
thres = np.copy(wb)
thres[thres > 95] = 255
thres[thres < 255] = 0

sns.heatmap(thres, annot=True, fmt="d", cmap="gray", cbar=False,
           xticklabels=False, yticklabels=False)
```

```python
kernel = np.ones((3,3),np.uint8)

erosion_dst = cv2.erode(thres,kernel)

sns.heatmap(erosion_dst, annot=True, fmt="d", cmap="gray", cbar=False,
           xticklabels=False, yticklabels=False)
```

```python
kernel = np.ones((2,2),np.uint8)

dilate_dst = cv2.dilate(thres,kernel)

sns.heatmap(dilate_dst, annot=True, fmt="d", cmap="gray", cbar=False,
           xticklabels=False, yticklabels=False)
```

```python
invented = np.array([[100, 200, 250],
                     [5,    35, 255],
                     [180,  24,   0]])

sns.heatmap(invented, annot=True, fmt="d", cmap="gray", cbar=False,
           xticklabels=False, yticklabels=False)
```
